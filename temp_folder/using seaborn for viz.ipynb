```python
import seaborn as sns
import pandas as pd

# Load the restaurant tips dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Set up nice styling
sns.set_style("whitegrid")
sns.set_context("notebook")
```

```python
# Explore the data
print("Dataset shape:", tips.shape)
print("\nColumn names:", tips.columns.tolist())

print("\nFirst 5 rows:")
print(tips.head())

print("\nData types:")
print(tips.dtypes)

print("\nBasic statistics:")
print(tips.describe())
```

```python
# Step 2: Distribution Analysis (Histograms)
# Create a histogram of tip amounts

sns.histplot(data = tips, x = 'tip', bins = 10, hue = 'time')
```

```python
# Create a histogram of total bill amounts  
# Compare tip distributions by time of day
sns.histplot(data = tips, x = 'total_bill', bins = 10, hue = 'time', palette = 'Set1')
```

```python
# Step 3: Relationship Analysis (Scatter Plots)

# Explore the relationship between bill and tip 
# Add a trend line to see the relationship clearly 
# Color the points by time of day 

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# regplot version
sns.regplot(data = tips, x = 'total_bill', y = 'tip', scatter = True)
```

```python
# lmplot
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time')
```
